## Final code for geocoding

```python
for index,name in enumerate(set(school_data['school_name'])):
    url= "https://developers.onemap.sg/commonapi/search?returnGeom=Y&getAddrDetails=Y&pageNum=1&searchVal="+name
    print(index,url)
    response = requests.get(url)
    data = json.loads(response.text) 
    
    temp_df = pd.DataFrame.from_dict(data['results'])
    temp_df['school_name'] = name
    
    if index == 0:
        school_geoloc = temp_df
    else:
        school_geoloc = school_geoloc.append(temp_df)

```

## Get distance between all addresses with the schools

```python
schools = pd.read_csv('C:/Users/User/Desktop/TP Y2(Sem2)/Data Mining & Business Analytics/Individual Assignment/school_with_latlon.csv')
```

import csv containing all addresses and respective lat & long

```python
hdb_addresses = pd.read_csv("C:/Users/User/Desktop/TP Y2(Sem2)/Data Mining & Business Analytics/Individual Assignment/Dataset(transformed)/address-2017-2021-csv.csv")
```

```python
hdb_addresses.head()
```

```python
print("num of duplicated rows:", hdb_addresses.duplicated().sum())
# drop duplicate address
# addresses are duplicated over the years 
hdb_addresses = hdb_addresses.drop_duplicates().reset_index() 
print("num of duplicated rows after dropping:", hdb_addresses.duplicated().sum())
```

```python
# index column is present thus will drop 
hdb_addresses = hdb_addresses.drop(columns=["index"])
```

```python
# make sure the sequence of the columns are correct
hdb_addresses.head()
```

```python
schools.head()
```

```python
from geopy.distance import geodesic

# make sure that 1st column is schoolname, 2nd = latitude and 3rd = longitude
school_temp_df = schools[['school_name','LATITUDE','LONGITUDE']]
flat_sch_distance = []

for index,flat in enumerate(hdb_addresses.iloc[:,0]):
    
    # 2nd column must be latitude, 3rd column must be longitude
    flat_loc = (hdb_addresses.iloc[index,1],hdb_addresses.iloc[index,2])
    # [ind,1] = latitude, [ind,2] = longitude
    for ind, sch in enumerate(school_temp_df.iloc[:,0]):
        sch_loc = (school_temp_df.iloc[ind,1],school_temp_df.iloc[ind,2])
        distance = geodesic(flat_loc,sch_loc)
        
        flat_sch_distance.append((flat,sch,distance.km))

flat_sch_distance_df = pd.DataFrame(flat_sch_distance)
flat_sch_distance_df.columns = ['address','school_name','distance_km']
flat_sch_distance_df.shape
```

```python
# export to csv as checkpoint data
flat_sch_distance_df.to_csv("hdb_flat_sch_distance_df.csv")
```
